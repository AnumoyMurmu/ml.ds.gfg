```python
import pandas as pd
```

```python
data=pd.read_csv('googleplaystore.csv')
```

```python
df=data.dropna()
```

```python
df
```

### total number of apps in each category:

```python
categories={}

for name in df['Category'].unique():
    ct=0
    for i in df['Category']:
        if(i==name):
            ct += 1
    print(name,":",ct)
    
    categories[name]=ct
```
